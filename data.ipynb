```python
import pandas as pd
df=pd.read_csv('data.csv')
df
```

```python
df['Name']=df['Name'].str.title()
df
```

```python
df['Gender']=df['Gender'].replace({'Male':'m','Female':'F','female':'F'})
df
```

```python
df['Gender']=df['Gender'].str.title()
df
```

```python
df['City'] = df['City'].str.strip().str.title()

# 2. Fix 'JoinDate' format and handle "not available"
df['JoinDate'] = pd.to_datetime(df['JoinDate'], errors='coerce')
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Name'] = df['Name'].fillna('Unknown')
df['City'] = df['City'].fillna('Unknown')
df
```

```python

df = df.drop_duplicates(subset=['Name'])
df
```

```python
df = df.drop_duplicates(subset=['Name'], keep='first')
df
```

```python
df = df.drop(index=[2, 5])
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df['JoinDate'] = df['JoinDate'].fillna('Not Available')
df['City'] = df['City'].replace('Unknown', 'Not Provided')
df['Salary'] = df['Salary'].astype(int)
df
```

```python
df = df[['Name', 'Gender', 'Age', 'Salary', 'JoinDate', 'City']]
df
```

```python
df.to_csv("Cleaned_Employee_Data.csv", index=False)
df
```
